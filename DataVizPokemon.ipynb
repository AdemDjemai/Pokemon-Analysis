##### Display pie chart of each type's frequency

```python
# Define a dictionary with logical colors for each type
type_colors = {
    'Normal': '#A8A77A',
    'Fire': '#EE8130',
    'Water': '#6390F0',
    'Electric': '#F7D02C',
    'Grass': '#7AC74C',
    'Ice': '#96D9D6',
    'Fighting': '#C22E28',
    'Poison': '#A33EA1',
    'Ground': '#E2BF65',
    'Flying': '#A98FF3',
    'Psychic': '#F95587',
    'Bug': '#A6B91A',
    'Rock': '#B6A136',
    'Ghost': '#735797',
    'Dragon': '#6F35FC',
    'Dark': '#705746',
    'Steel': '#B7B7CE',
    'Fairy': '#D685AD'
}
# Types counts
type1_counts = df['Type'].value_counts()
type2_counts = df['Other Type'].value_counts()
```

```python
# Plot the pie chart of each type1's frequency
plt.figure(figsize=(8, 8))
plt.pie(type1_counts, labels=type1_counts.index, autopct='%1.1f%%',
        colors=[type_colors[i] for i in type1_counts.index])
plt.title('Type 1 Distribution')
plt.axis('equal')
plt.show()
```

```python
# Plot the pie chart of each type2's frequency
plt.figure(figsize=(8, 8))
plt.pie(type2_counts, labels=type2_counts.index, autopct='%1.1f%%',
        colors=[type_colors[j] for j in type2_counts.index])
plt.title('Type 2 Distribution')
plt.axis('equal')
plt.show()
```

##### Calculate the count of Pokemon in each generation

```python
# Calculate the generation counts
generation_counts = df['Generation'].value_counts().sort_index()

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(8, 6))

# Create the bar chart
bar_colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightpink', 'lightyellow', 'lightsalmon']
ax.bar(generation_counts.index, generation_counts.values, color=bar_colors)

# Add labels and title
ax.set_xlabel('Generation', fontsize=12)
ax.set_ylabel('Number of Pokemon', fontsize=12)
ax.set_title('Generation Pokemon Count', fontsize=14, fontweight='bold')

# Add grid lines
ax.grid(axis='y', linestyle='--', alpha=0.5)

# Customize tick labels
ax.tick_params(axis='x', labelsize=10)

# Add data labels
for i, count in enumerate(generation_counts.values):
    ax.text(generation_counts.index[i], count + 5, str(count), ha='center', fontsize=8)

# Adjust spacing and margins
plt.tight_layout()

# Display the chart
plt.show()
```

##### Types Matchup Chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Feature Engineering
type_chart = pd.read_csv('types_chart.csv', index_col=0)

types = list(type_chart.columns)
chart_data = type_chart.values.astype(float)

fig, ax = plt.subplots(figsize=(10, 10))
heatmap = ax.imshow(chart_data, cmap='RdBu_r', vmin=0.5, vmax=2)

ax.set_xticks(np.arange(len(types)))
ax.set_yticks(np.arange(len(types)))
ax.set_xticklabels(types, fontsize=10)
ax.set_yticklabels(types, fontsize=10)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right")
plt.colorbar(heatmap, fraction=0.046, pad=0.04)
plt.title("Pokemon Type Matchup Chart", fontsize=14, fontweight='bold')
plt.xlabel("Defending Type", fontsize=12)
plt.ylabel("Attacking Type", fontsize=12)

for i in range(len(types)):
    for j in range(len(types)):
        text = ax.text(j, i, chart_data[i, j], ha="center", va="center", color="black", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
type_chart = pd.read_csv('types_chart.csv', index_col=0)

# Generate all combinations of types
types = type_chart.columns.tolist()
type_combinations = [(t1, t2) for t1 in types for t2 in types if t1 != t2]

# Create a dataframe to store the type combinations and their scores
df_combinations = pd.DataFrame(type_combinations, columns=['Type', 'Other Type'])
df_combinations['Offense'] = df_combinations.apply(lambda row: type_chart.loc[row['Type']].sum() + type_chart.loc[row['Other Type']].sum(), axis=1)
df_combinations['Defense'] = df_combinations.apply(lambda row: type_chart[row['Type']].sum() + type_chart[row['Other Type']].sum(), axis=1)

# Calculate the min and max defense values
max_defense = df_combinations['Defense'].max()

# Calculate the defense grade for each type combination
df_combinations['Defense Grade'] = df_combinations['Defense'].apply(lambda defense: (defense / max_defense) * 20)

# Sort the dataframe in ascending order based on 'value'
df_combinations = df_combinations.sort_values('Defense')

# Calculate the min and max defense grades
min_dgrade = df_combinations['Defense Grade'].min()
max_dgrade = df_combinations['Defense Grade'].max()

# Flip the values
df_combinations['Defense Grade'] = max_dgrade - df_combinations['Defense Grade'] + min_dgrade

# Calculate the max offense value
max_offense = df_combinations['Offense'].max()

# Calculate the offense grade for each type combination
df_combinations['Offense Grade'] = df_combinations['Offense'].apply(lambda offense: (offense / max_offense) * 20)


# Calculate the average grade
df_combinations['Average Grade'] = df_combinations[['Defense Grade', 'Offense Grade']].mean(axis=1)

# Remove direct duplications
df_combinations = df_combinations.drop_duplicates(subset=['Type', 'Other Type'])

# Remove alternate duplications
df_combinations_alt = df_combinations.copy()
df_combinations_alt[['Type', 'Other Type']] = df_combinations[['Other Type', 'Type']]
df_combinations = pd.concat([df_combinations, df_combinations_alt]).drop_duplicates()


# Sort the combinations based on the average grade
df_combinations = df_combinations.sort_values(by='Average Grade', ascending=False)

# Display the top combinations
top_combinations = df_combinations.head(20)
top_combinations_list = top_combinations.apply(lambda row: f"{row['Type']}/{row['Other Type']}", axis=1).tolist()

# Create a bar chart to visualize the rankings
plt.figure(figsize=(10, 6))
plt.barh(range(len(top_combinations)), top_combinations['Average Grade'], color='steelblue')
plt.yticks(range(len(top_combinations)), top_combinations_list)
plt.xlabel('Average Grade')
plt.ylabel('Type Combinations')
plt.title('Top Type Combinations Ranked by Average Grade')
plt.gca().invert_yaxis()
plt.show()


```
